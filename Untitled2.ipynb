```python
from matplotlib.colors import ListedColormap
from sklearn import model_selection, datasets, linear_model, metrics

import numpy as np
```

```python
%pylab inline
```

```python
blobs = datasets.make_blobs(centers=2, cluster_std=5.5, random_state=42)
```

```python
blobs
```

```python
colors = ListedColormap(['red', 'blue'])

pylab.figure(figsize(8,8))
pylab.scatter(list(map(lambda x: x[0], blobs[0])), list(map(lambda x:x[1], blobs[0])), c=blobs[1], cmap=colors)
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(blobs[0], blobs[1], test_size=0.3, random_state=42)
```

```python
r_Classifier = linear_model.RidgeClassifier()
```

```python
r_Classifier.fit(train_data, train_labels)
```

```python
r_pred = r_Classifier.predict(test_data)
```

```python
r_pred.shape
```

```python
a=0
for i in range(len(r_pred)):
    if r_pred[i]==test_labels[i]:
        a+=1
print(a)
    
```

```python
metrics.accuracy_score(test_labels, r_pred)
```

```python
r_Classifier.coef_
```

```python
log_reg = linear_model.LogisticRegression(random_state=42)
log_reg.fit(train_data, train_labels)
```

```python
lg_pred = log_reg.predict(test_data)
```

```python
lg_pred_proba = log_reg.predict_proba(test_data)
```

```python
metrics.accuracy_score(test_labels, lg_pred)
```

```python
lg_pred_proba
```

```python
ridge_scor = model_selection.cross_val_score(r_Classifier, blobs[0], blobs[1], scoring='accuracy', cv=10)
lg_scor = model_selection.cross_val_score(log_reg, blobs[0], blobs[1], scoring='accuracy', cv=10)
```
