```python
import pandas as pd
```

```python
money = pd.read_excel("ru.xlsx")
money
```

```python
money.curs
```

```python
money[:29:-1]
```

```python
money[:516-29:-1]
```

```python
past = 7 * 4 # на основе данных за 4 недели 
             #будем предсказывать на неделю вперед
```

```python
future = 7 
```

```python
values = money.curs
```

```python
start = past
```

```python
end = len(values) - future 
```

```python
end
```

```python
start
```

```python
raw_df = []
for i in range(start, end):
        p_a_f = values[(i-past):(i+future)]
        print(p_a_f)
        raw_df.append(list(p_a_f))
print(raw_df)
```

```python
raw_df
```

```python
past_col = [f"past_{i}" for i in range(past)]
```

```python
past_col
```

```python
fut_col = [f"future_{i}" for i in range(future)]
```

```python
fut_col
```

```python
df = pd.DataFrame(raw_df, columns=(past_col + fut_col))
```

```python
df
```

```python
x = df[past_col][:-1]
x 
```

```python
y = df[fut_col][:-1]
```

```python
y
```

```python
x_test = df[past_col][-1:]
x_test
```

```python
y_test = df[fut_col][-1:]
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LinReg = LinearRegression()
LinReg.fit(x,y)
```

```python
pred = LinReg.predict(x_test)
```

```python
pred
```

```python
y_test
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(pred[0], label="Prediction")
plt.plot(y_test.iloc[0], label="Real Data")
plt.legend()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(pred[0], y_test.iloc[0])
```

```python
MLP
```

```python
pred = MLP.predict(x_test)
```
